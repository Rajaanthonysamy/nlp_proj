```python
import nltk
from nltk.tokenize import word_tokenize

##Punkt is a pretrained sentence tokenizer used in Natural Language Processing (NLP).
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
with open('sherlock-holm.es_stories_plain-text_advs.txt', 'r', encoding='utf-8') as f:
    text = f.read().lower()
```

```python
text
```

```python
token_data=word_tokenize(text)
```

```python
token_data
```

```python
len(token_data)
```

```python
from collections import Counter

word_count=Counter(token_data)
```

```python
word_count
```

```python
vocab= sorted(word_count, key=word_count.get, reverse=True)
```
